```python
train_rv = rv.iloc[0:450]
test_rv = rv.iloc[450::]
```

```python
train_rv.shape
```

```python
test_rv.shape
# 4 weeks data in test data
```

```python
import pmdarima
from pmdarima import auto_arima
```

```python
auto_arima(rv.total, seasonal= True, m= 7)
```

```python
# whatever value we get in auto_arimaa , we pass it in SARIMA
from statsmodels.tsa.statespace.sarimax import SARIMAX
```

```python
model1 = SARIMAX(train_rv.total, order=(0, 0, 0),seasonal_order=(1, 0, 1, 7) )
```

```python
results = model1.fit()
```

```python
pred_rv = results.predict(start = 450, end = 477)
pred_rv
```

```python
# Lets check if SARIMA model is able to handle the holidays

test_rv[test_rv.holiday == 1]
```

```python
plt.figure(figsize= (8,8))
plt.plot(test_rv.total, marker = '*', color = 'green', label = 'Actual')
plt.plot(pred_rv, marker = '*', color = 'red', label = 'Predicted')
plt.legend()

plt.axvline(x = '2017-04-14', color = 'yellow' )
plt.axvline(x = '2017-04-16', color = 'yellow' )
plt.axvline(x = '2017-04-17', color = 'yellow' )
```

```python
# We can see that SARIMA model is not doing good for holidays
```

```python
# Lets find the error

error = test_rv.total - pred_rv
```

```python
MSE_sarima = np.mean(np.square(error))
MSE_sarima
```
