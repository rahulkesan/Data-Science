## Has the model OVERFITTED?

A. MSE in train data is almost zero in front of MSE in test data <br>
   Same goes with MAPE.<br>
   This give us the conclusion that model is doing very good with train data but heavily failing in test data<br>
   Also, our target variable is in 3 digit but the MSE in in 17 digit.<br>
   
#### Hence, we conclude that model is Overfitted

```python

from sklearn.linear_model import Ridge
```

```python
ridreg = Ridge()
```

```python
ridreg.fit(tr_train_x, tr_train_y)
```

```python
# Now checking our parameters with Ridge

pred_train_rig = ridreg.predict(tr_train_x)
error_train_rig = tr_train_y - pred_train_rig
mse_train_rig = np.mean(np.square(error_train_rig))
mape_train_rig = np.mean(np.absolute(error_train_rig*100/tr_train_y))



pred_test_rig = ridreg.predict(tr_test_x)
error_test_rig = tr_test_y - pred_test_rig
mse_test_rig = np.mean(np.square(error_test_rig))
mape_test_rig = np.mean(np.absolute(error_test_rig*100/tr_test_y))
```

```python
print(mse_train_rig, mse_test_rig)
```

```python
print(mape_train_rig, ',', mape_test_rig)
```

```python
# Now, our Ridge model has taken care of overfitting
```

```python
# Similarly we will do with Lasso

from sklearn.linear_model import Lasso
```

```python
lasreg = Lasso()
```

```python
lasreg.fit(tr_train_x, tr_train_y)
```

```python
# Now checking our parameters with Lasso

pred_train_las = lasreg.predict(tr_train_x)
error_train_las = tr_train_y - pred_train_las
mse_train_las = np.mean(np.square(error_train_las))
mape_train_las = np.mean(np.absolute(error_train_las*100/tr_train_y))



pred_test_las = lasreg.predict(tr_test_x)
error_test_las = tr_test_y - pred_test_las
mse_test_las = np.mean(np.square(error_test_las))
mape_test_las = np.mean(np.absolute(error_test_las*100/tr_test_y))
```

```python
print(mse_train_las, ',', mse_test_las)
```

```python
print(mape_train_las, ',', mape_test_las)
```
