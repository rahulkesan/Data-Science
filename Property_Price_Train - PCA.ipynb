```python
from sklearn import decomposition
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# Lets Standardise our data --> x-mu/sd

scaler = StandardScaler()
```

```python
pa2 = pa1.sample(frac= 1, axis= 1)
```

```python
pa2 = pa2.drop(['Sale_Price'], axis =1)
```

```python
pa2.head()
```

```python
scaled_pa = scaler.fit_transform(pa)
```

```python
scaled_pa
```

```python
# PCA Transformtion

pca_pa = PCA()
```

```python
pca_x = pca_pa.fit_transform(scaled_pa)

# PCA transformation is done
```

```python
scaled_pa.shape
```

```python
pca_x.shape  # Has dimensionality reduction taken place?? -- > No
```

```python
pca_pa.explained_variance_ratio_
# Anything special observed in these number
```

```python
list(pca_pa.explained_variance_ratio_)

# We can see that the numbers are positive and are in descending order as in they are decreasing
# all the values are less than 1
```

```python
pca_pa.explained_variance_ratio_.sum()

# This sum should come down to be 1
```

```python
len(pca_pa.explained_variance_ratio_)

# For each column, one principal component has been created
```

```python
# Any model we assume that there should be 5% to 10% of error

# Lets select those many principal components which are able to explain around 90% to 95% of the variance(Rsquare)
# And we can ignore the rest
```

```python
l1 = list(pca_pa.explained_variance_ratio_)
```

```python
np.sum(l1[0:55])

# Around 1st 55 PC are able to explain around 95% of the variance

# By doing this we have done the dimensionality reduction
```

```python
pca_x # data after pca transformation
```

```python
new_data = pca_x[:, 0:2]
```

```python
new_data.shape
```

```python
# Now we will build linear regression model using only 55 components and not all the components
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
```

```python
lm.fit(new_data, pa1.Sale_Price)
```

```python
lm.score(new_data, pa1.Sale_Price)
```

```python
# when 55 PC variable Rsquare --> .8484359
# when 65 PC variable Rsquare --> .8518202
# when 70 PC variable Rsquare --> .8537659
# When taking all the component except 1st one Rsquare --> 0.121301
# When taking 1st two component Rsquare --> 0.7407888
```

```python
# 1st principal component is always the most imp, followed by 2nd, then 3rd and so on
```

```python
# PCA is used for dimensionality reduction but not for feature selection.
# Because we do not know which variable is most significant
# This is because their is no one on one mapping between original columns and Principal components
```

```python
# PCA is used to overcome the problem of multicolinearity
```
