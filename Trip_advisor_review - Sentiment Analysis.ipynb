```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

# Vader

```python
nltk.download('vader_lexicon')
```

```python
sent = SentimentIntensityAnalyzer()
# Creating object sent
```

```python
sent.polarity_scores("juice was awesome and i got heavenly feeling after taking a sip")
```

```python
sent.polarity_scores("journey was horrible and was late")
```

```python
# Value of compound will be from -1 to +1
# Where -ve indicate negative sentiment and +ve indicate positive sentiment
```

```python
sent.polarity_scores("juice was awesome and i got heavenly feeling after taking a sip and the food was awesome")
```

```python
sent.polarity_scores("movie was good and I enjoyed sleeping")
```

```python
ta.head()
```

```python
# Consider rating column is not there and we have to find if overall sentiment is positive or negative
```

```python
out_score = sent.polarity_scores("juice was awesome and i got heavenly feeling after taking a sip")
```

```python
out_score
```

```python
out_score['compound']
```

```python
# Now lets run the entire logic on the trip advisor data
```

```python
score = []
```

```python
ta.head()
```

```python
for comment in ta['Review']:
    var1 = sent.polarity_scores(comment)
    score.append(var1['compound'])
```

```python
score
```

```python
ta['Score'] = score
```

```python
ta.head()
```

```python
ta.Score.describe(percentiles= [.1,.2,.3,.4,.5,.6,.7,.8,.9,1])
```

```python
ta[ta.Score < 0]
```

```python
# 1617 reviews are negative and most are positive
```

```python
# lets run the entire logic on the trip advisor data using textblob
```

```python
ta.head()
```

```python
ta = ta.iloc[:, 0:-1]
```

```python
ta.head()
```

```python
score_blob = []
for comment in ta['Review']:
    var2 = TextBlob(comment)
    score_blob.append(var2.sentiment[0])
```

```python
score_blob
```

```python
len(score_blob)
```

```python
ta['Score_Blob'] = score_blob
```

```python
ta.head()
```

```python
ta.Score_Blob.describe(percentiles=[.1,.2,.3,.4,.5,.6,.7,.8,.9,1])
```

```python
ta[ta.Score_Blob > 0] # positive
```

```python
ta[ta.Score_Blob < 0] # negative
```
