```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
from sklearn.tree import DecisionTreeClassifier
```

```python
ch_train, ch_test = train_test_split(ch, test_size= .2)
```

```python
ch_train_x = ch_train.iloc[:, 0:-1]
ch_train_y = ch_train.iloc[:, -1]

ch_test_x = ch_test.iloc[:, 0:-1]
ch_test_y = ch_test.iloc[:, -1]
```

```python
logreg = LogisticRegression()
dt = DecisionTreeClassifier()
```

```python
logreg.fit(ch_train_x, ch_train_y)
```

```python
dt.fit(ch_train_x, ch_train_y)
```

```python
pred_logreg = logreg.predict(ch_test_x)
pred_dt = dt.predict(ch_test_x)
```

```python
tab_logreg = confusion_matrix(ch_test_y, pred_logreg)
tab_dt = confusion_matrix(ch_test_y, pred_dt)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
print('---FROM LOGISTIC REGRESSION MODEL---')

print('Accuracy score is --> ', accuracy_score(ch_test_y, pred_logreg)*100)
print('Precision score is --> ', precision_score(ch_test_y, pred_logreg)*100)
print('Recall score is --> ', recall_score(ch_test_y, pred_logreg)*100)
print('F1 score is --> ', f1_score(ch_test_y, pred_logreg)*100)
```

```python
print('---FROM DECISION TREE MODEL---')

print('Accuracy score is --> ', accuracy_score(ch_test_y, pred_dt)*100)
print('Precision score is --> ', precision_score(ch_test_y, pred_dt)*100)
print('Recall score is --> ', recall_score(ch_test_y, pred_dt)*100)
print('F1 score is --> ', f1_score(ch_test_y, pred_dt)*100)
```
